```python
def LatexCompiler(Array2D, Print):
    newarr = np.zeros((int(Array2D.shape[0]+1), int(Array2D.shape[1]+1)))
    newarr[:-1,0] = np.arange(1,Array2D.shape[0]+1)
    newarr[:-1,1:] = Array2D
    newarr[-1,1:] = np.mean(Array2D, axis = 0)
    b = '\\\\'.join('&'.join('%0.2f' %x for x in y) for y in newarr)
    
    if Print == True:
        print(newarr)
    return b
```

```python
import numpy as np
import tensorflow as tf
import tensorflow.keras as ks
from numpy import hstack
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Conv1D
from tensorflow.keras.layers import MaxPooling1D
from tensorflow.keras.layers import concatenate
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
train_idx = 12000 # 80% of data to

from numpy import load
# load array
X1 = load('Train1.npy')
X2 = load('Train2.npy')
y = load('Train_targets.npy')



X1FFN = np.reshape(X1,(X1.shape[0],X1.shape[1]*X1.shape[2]))
Xcombo = np.concatenate((X1FFN,X2), axis = 1)



# RANDOMISING DATA
###############################################################
X1LSTM = np.transpose(X1, [0,2,1])
randind = np.arange(len(y))
np.random.shuffle(randind)
tempXc = Xcombo[randind]
tempX1LSTM = X1LSTM[randind]
tempX1 = X1[randind]
tempX2 = X2[randind]
tempy1 = y[randind]


# Training Data
xc = tempXc[0:train_idx]
x1lstm = tempX1LSTM[0:train_idx]
x1 = tempX1[0:train_idx]
x2 = tempX2[0:train_idx]
y1 = tempy1[0:train_idx]


# Testing Data
testXc = tempXc[train_idx::]
testXLSTM = tempX1LSTM[train_idx::]
testX1 = tempX1[train_idx::]
testX2 = tempX2[train_idx::]
testy = tempy1[train_idx::]
################################################################
```

```python
# FFNN
def ModelPerformance(yp, yt):
    G1Tot = 0
    G1Correct = 0
    G2Tot = 0
    G2Correct = 0
    G3Tot = 0
    G3Correct = 0
    G4Tot = 0
    G4Correct = 0
    G5Tot = 0
    G5Correct = 0
    G6Tot = 0
    G6Correct = 0
    
    for i in range(len(yp)):
        if yt[i] <= 12:
            G1Tot += 1
            if yp[i] <= 12:
                G1Correct +=1
        elif yt[i] <= 35:
            G2Tot += 1
            if yp[i] > 12 and yp[i] <= 35:
                G2Correct += 1
        elif yt[i] <= 55:
            G3Tot += 1
            if yp[i] > 35 and yp[i] <= 55:
                G3Correct += 1
        elif yt[i] <= 150:
            G4Tot += 1
            if yp[i] > 55 and yp[i] <= 150:
                G4Correct += 1
        elif yt[i] <= 250:
            G5Tot += 1
            if yp[i] > 150 and yp[i] <= 250:
                G5Correct += 1
        elif yt[i] >250:
            G6Tot += 1
            if yp[i] > 250:
                G6Correct += 1
                
    MSE   = np.mean((yp-yt)**2)
    RMSE  = (MSE)**0.5
    MAPE  = np.mean(np.abs((yt-yp)/(yt)))
    OBAR  = np.mean(yt)
    IA    = 1 - np.sum((yp-yt)**2)/np.sum((np.abs(yp-OBAR) + np.abs(yt-OBAR))**2)
    G1acc = G1Correct/G1Tot * 100
    G2acc = G2Correct/G2Tot * 100
    G3acc = G3Correct/G3Tot * 100
    G4acc = G4Correct/G4Tot * 100
    G5acc = G5Correct/G5Tot * 100
    G6acc = G6Correct/G6Tot * 100
    Wacc  = ((G1acc*G1Correct)+(G2acc*G2Correct)+(G3acc*G3Correct)+(G4acc*G4Correct)+(G5acc*G5Correct)+(G6acc*G6Correct))/(len(yp))
    
    return [G1acc, G2acc, G3acc, G4acc, G5acc, G6acc, Wacc]


def build_FFNN(Xcombo):
    ffnnx = Input(shape=(Xcombo.shape[1]))
    cnnx = Dense(200, activation = 'swish')(ffnnx)
    cnnx = Dense(100, activation = 'swish')(cnnx)
    cnnx = Dense(50, activation = 'swish')(cnnx)
    cnnx = Dense(10, activation = 'swish')(cnnx)
    outputx = Dense(1)(cnnx)
    
#     modelX = Model(inputs=[ffnnx1,ffnnx2], outputs=outputx)
    modelX = Model(inputs=[ffnnx], outputs=outputx)
    modelX.compile(optimizer='adam', loss='mse', metrics = ['MeanAbsolutePercentageError', 'RootMeanSquaredError'])
    return modelX


def build_CNN(p_size, strd, act, dpt, X1, X2):
    in1 = Input(shape=(X1.shape[1], X1.shape[2]))

    cnnx = Conv1D(filters=16, kernel_size= 50, strides = strd, activation=act)(in1)
    cnnx = Dropout(dpt)(cnnx) 

    cnnx = BatchNormalization()(cnnx)
    cnnx = Conv1D(filters=20, kernel_size= 30, strides = strd, activation=act)(cnnx)
    cnnx = Dropout(dpt)(cnnx) 

    cnnx = BatchNormalization()(cnnx)
    cnnx = Conv1D(filters=32, kernel_size= 16, strides = strd, activation=act)(cnnx)
    cnnx = Dropout(dpt)(cnnx) 

    cnnx = BatchNormalization()(cnnx)
    cnnx = Conv1D(filters=64, kernel_size=  8, strides = strd, activation=act)(cnnx)
    cnnx = MaxPooling1D(pool_size=p_size)(cnnx)
    cnnx = Dropout(dpt)(cnnx)

    cnnx = BatchNormalization()(cnnx)
    cnnx = Conv1D(filters=128, kernel_size= 4, strides = strd, activation=act)(cnnx)
    cnnx = MaxPooling1D(pool_size=p_size)(cnnx)
    cnnx = Dropout(dpt)(cnnx)

    cnnx = BatchNormalization()(cnnx)
    cnnx = Conv1D(filters=256, kernel_size= 2, strides = strd, activation=act)(cnnx)
    cnnx = MaxPooling1D(pool_size=p_size)(cnnx)
    cnnx = Dropout(dpt)(cnnx)

    cnnx = Flatten()(cnnx)

    in2 = Input(shape=(X2.shape[1]))
    ffnnx = Dense(50, activation = 'sigmoid')(in2)

    mergex = concatenate([cnnx,ffnnx])
    densex = Dense(30, activation=act)(mergex)
    outputx = Dense(1)(densex)

    modelX = Model(inputs=[in1, in2], outputs=outputx)
    modelX.compile(optimizer='adam', loss='mse', metrics = ['MeanAbsolutePercentageError', 'RootMeanSquaredError'])
    return modelX


def build_LSTM(act, dpt, X1, X2):
    in1 = Input(shape=(X1.shape[-2:]))
    LSTMx = LSTM(units = 150,  dropout = dpt)(in1)

    in2 = Input(shape=(X2.shape[1]))
    ffnnx = Dense(50, activation = 'sigmoid')(in2)

    mergex = concatenate([LSTMx,ffnnx])
    densex = Dense(30, activation=act)(mergex)
    outputx = Dense(1)(densex)
    modelX = Model(inputs=[in1, in2], outputs=outputx)
    modelX.compile(optimizer='adam', loss='mse', metrics = ['MeanAbsolutePercentageError', 'RootMeanSquaredError'])
    return modelX
```

```python
EPOCHS = 1000
BATCH = 256
early_stop = ks.callbacks.EarlyStopping(monitor='val_loss', patience=30,restore_best_weights = True, 
                                            min_delta = 5, verbose=2)

# Create Models
model_FFNN = build_FFNN(Xcombo)
model_CNN = build_CNN(2, 1, 'swish', 0.1, X1, X2)
model_LSTM = build_LSTM('swish', 0.2, X1LSTM, X2)

# Fit Model
mFFNN_hist = model_FFNN.fit([xc], y1, epochs=EPOCHS,validation_split=0.15, verbose=2, callbacks = [early_stop], batch_size=BATCH, shuffle = True)
mCNN_hist = model_CNN.fit([x1,x2], y1, epochs=EPOCHS,validation_split=0.15, verbose=2, callbacks = [early_stop], batch_size=BATCH, shuffle = True)
mLSTM_hist = model_LSTM.fit([x1lstm,x2], y1, epochs=EPOCHS,validation_split=0.15, verbose=2, callbacks = [early_stop], batch_size=64, shuffle = True)
# Evaluate Model
y_predict_FFNN = model_FFNN.predict([testXc])
y_predict_CNN = model_CNN.predict([testX1,testX2])
y_predict_LSTM = model_LSTM.predict([testXLSTM,testX2])
```
